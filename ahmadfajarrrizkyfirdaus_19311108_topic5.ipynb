```python
import matplotlib.pyplot as plt
import numpy as np
plt.boxplot(df['tenure'])
plt.show()
```

```python
Q1= np.percentile(df.tenure,25)
Q3= np.percentile(df.tenure,75)
IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
df[df['tenure']>upper_bound]
```

```python
outliers_tenure_df = df[df['tenure']>upper_bound]
outliers_tenure_df 
```

```python
df.loc[(df['tenure']>upper_bound),'tenure'] = upper_bound
```
